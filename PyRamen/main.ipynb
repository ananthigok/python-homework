```python
#pyRamen-main.ipynb
import pathlib 
import csv
## function to write the output to file 
def writeOutputToFile(fileName, data):
    File_object = open(fileName, "w")
    File_object.writelines(data)
    File_object.close()
#function to read csv and returns contents as list 
def read_csv(csvFile):
    csvList = []
    with csvFile.open() as f:
        csvReader = csv.reader(f, delimiter=',', quotechar='"')
        counter = 0
        for row in csvReader:
            if (counter==0):
                next(csvReader)
            csvList = list(csvReader)
    return csvList

csvFileMenu = pathlib.Path('./menu_data.csv')
csvListMenu = read_csv(csvFileMenu)

csvFileSales = pathlib.Path('./sales_data.csv') 
csvListSales = read_csv(csvFileSales)

report = {}
sales_item = {}
counter=0
for row in csvListSales: 
    counter += counter
    #print (row) ## 1 => id, 2 => date, 3 => creditcard, 4 => quantity, 5 => menu 
    print ()
    item = row[4].lstrip()
    quantity = int(row[3])
    if item in sales_item:
        quantity += int(sales_item[item]["quantity"])
    sales_item[item] = {
        "menu-item" : item,
        "quantity" : quantity,
    }    
item = ''
quantity = 0
profit = 1
menu_item = []
counter = 0
for rowMenu in csvListMenu:
    #print (rowMenu) ## 1 => item, 2 => category, 3 => description, 4 => price, 5 => cost 
    item = rowMenu[0]
    price = rowMenu[-2]
    cost = rowMenu[-1]
    menu_item.append({
        "item" : item,
        "price" : price,
        "cost" : cost
    })
    profit = int(float(price))-int(float(cost))
    if item in sales_item:
        quantity = int(sales_item[item]['quantity'])
        revenue = int(price) * (quantity)
        cogs = int(cost) * (quantity)
        profit = profit * quantity
        if item in report:
            quantity += report[item]["01-count"]
            revenue += report[item]["02-revenue"]
            cogs += report[item]["03-cogs"]
            profit += report[item]["04-profit"]
        report[item]={
            "01-count" : quantity,
            "02-revenue" : revenue,
            "03-cogs" :  cogs,
            "04-profit" : float(profit)
          }   
    else:
        print()
        print (f"sales item does not equal {item} - NO MATCH")
print (report)
writeOutputToFile('pyRamen-report',str(report))
```
